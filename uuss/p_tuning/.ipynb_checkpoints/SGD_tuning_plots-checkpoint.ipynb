```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

'0' epoch is the pre-trained models performance on uuss validation dataset

```python
learning_rates = ["0.00005", "0.0001", "0.0005", "0.001", "0.005", "0.01", "0.05", "0.1"]
weight_decays = ["5e-5", "1e-4", "5e-4", "1e-3", "5e-3"]

for lr in learning_rates:
    for wd in weight_decays:
        dir_name = f"SGD_tuning/128_{lr}_{wd}_15"
        df = pd.read_csv(f"{dir_name}/train_metrics.csv")
        plt.plot(range(len(df)-1), df.va_rms[:-1], label=f"{lr}_{wd}")
plt.legend()
```

```python
learning_rates = ["0.00005", "0.0001", "0.0005", "0.001", "0.005", "0.01"] #, "0.05", "0.1"]
weight_decays = ["5e-5", "1e-4", "5e-4", "1e-3", "5e-3"]

for lr in learning_rates:
    plt.figure()
    for wd in weight_decays:
        dir_name = f"SGD_tuning/128_{lr}_{wd}_15"
        df = pd.read_csv(f"{dir_name}/train_metrics.csv")
        plt.plot(range(len(df)), df.va_rms[:], label=f"{lr}_{wd}")
    plt.legend()
    plt.ylim([0.084, 0.1])
```
