```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_name = f"SWAG_tune_lr/128_0.003_5e-3_0.001_85_25_20/val_results_100/val_eval_swag-56_residual_summary_100.csv"
# colors = ["#9ebcda", "#8856a7"]
colors = ["k", "grey"]
# start at 3 samples
startN=3
df = pd.read_csv(file_name)
df = df.iloc[startN-1:]
fig, axes = plt.subplots(1, figsize=(5 ,4))
ln1 = axes.plot(np.arange(startN, len(df)+startN)+1, df.ens_of_std, color=colors[0], label="OFSTD")
axes1 = axes.twinx()
ln2 = axes1.plot(np.arange(startN, len(df)+startN)+1, df.ens_rms, color=colors[1], label="RMS")
axes.set_xlabel("N Samples", fontsize=12)
axes.set_ylabel("OFSTD of N samples", color=colors[0], fontsize=12)
axes1.set_ylabel("RMS of N samples", x=1.1, fontsize=12) #, color=colors[1])
lns = ln1 + ln2 
labs = [l.get_label() for l in lns]
axes.legend(lns, labs, fontsize=12)
# axes.set_yticks(np.arange(0.047, 0.057, 0.0020));
axes1.set_yticks(np.arange(0.1, 0.17, 0.02));
```

```python
file_name_p = f"../p_tuning/SWAG_tune_lr_ft24/128_0.0005_5e-3_0.0002_84_24_20/val_results_100/val_eval_swag-59_residual_summary_100.csv"
file_name_s = f"SWAG_tune_lr/128_0.003_5e-3_0.001_85_25_20/val_results_100/val_eval_swag-56_residual_summary_100.csv"
# colors = ["#9ebcda", "#8856a7"]
colors = ["k", "grey"]
# start at 3 samples
startN=5
dfP = pd.read_csv(file_name_p)
dfP = dfP.iloc[startN-1:]
dfS = pd.read_csv(file_name_s)
dfS = dfS.iloc[startN-1:]
fig, axes = plt.subplots(1, figsize=(5 ,4))
ln1 = axes.plot(np.arange(startN, len(dfP)+startN)+1, dfP.ens_rms, color=colors[0], label="P")
axes1 = axes.twinx()
ln2 = axes1.plot(np.arange(startN, len(dfS)+startN)+1, dfS.ens_rms, color=colors[1], label="S")
axes.set_xlabel("N Samples", fontsize=12)
axes.set_ylabel("P-arrival RMS (s)", color=colors[0], fontsize=12)
axes1.set_ylabel("S-arrival RMS (s)", x=1.1, fontsize=12) #, color=colors[1])
axes.set_title("Residual RMS for Mean of N Samples")
lns = ln1 + ln2 
labs = [l.get_label() for l in lns]
axes.legend(lns, labs, fontsize=12)
axes.set_yticks(np.arange(0.084, 0.088, 0.0010));
axes1.set_yticks(np.arange(0.1, 0.130, 0.01));
plt.savefig("PS_swag_rms_N_samples.png", dpi=300, bbox_inches='tight')
```
