```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Fine-tune RMS (where these start) is 0.08495

```python
learing_rates = ["0.00001", "0.0001", "0.001"] #, "0.01"] #, "0.1"]
for lr in learing_rates:
    dir_name = f"SWAG_tune_lr/128_0.003_5e-3_{lr}_60_25_20"
    df = pd.read_csv(f"{dir_name}/train_metrics.csv")
    plt.plot(range(len(df)), df.swa_va_rms, label=f"{lr}_swa")
plt.legend()
```

```python
learing_rates = ["0.0001", "0.0005", "0.0007", "0.001"] #, "0.002"] #, "0.01"] #, "0.1"]
for lr in learing_rates:
    dir_name = f"SWAG_tune_lr/128_0.003_5e-3_{lr}_60_25_20"
    df = pd.read_csv(f"{dir_name}/train_metrics.csv")
    plt.plot(range(len(df)), df.swa_va_rms, label=f"{lr}_swa")
plt.legend()
```

```python
dir_name = f"SWAG_tune_lr/128_0.003_5e-3_0.001_85_25_20"
df = pd.read_csv(f"{dir_name}/train_metrics.csv")
plt.plot(range(len(df)), df.swa_va_rms, label=f"0.001_swa")
```

```python
learing_rates = ["0.001"]
for lr in learing_rates:
    ens_rms = []
    for i in range(26, 85):
        file_name = f"SWAG_tune_lr/128_0.003_5e-3_{lr}_85_25_20/val_results_30/val_eval_swag-{i}_residual_summary_30.csv"
        df = pd.read_csv(file_name)
        ens_rms.append(df.ens_rms.iloc[29])
    plt.plot(range(len(ens_rms)), ens_rms, label=f"{lr}_swa")
        #plt.plot(range(len(df)), df.va_rms, label=f"{k}")
plt.legend()
plt.xlabel("Iterations")
plt.ylabel("RMS of 30 samples")
```

```python
learing_rates = ["0.001"]
for lr in learing_rates:
    ens_of_std = []
    for i in range(26, 85):
        file_name = f"SWAG_tune_lr/128_0.003_5e-3_{lr}_85_25_20/val_results_30/val_eval_swag-{i}_residual_summary_30.csv"
        df = pd.read_csv(file_name)
        ens_of_std.append(df.ens_of_std.iloc[29])
    plt.plot(range(len(ens_of_std)), ens_of_std, label=f"{lr}_swa")
        #plt.plot(range(len(df)), df.va_rms, label=f"{k}")
plt.legend()
plt.xlabel("Iterations")
plt.ylabel("OF STD of 30 samples")
```

```python
plt.figure()
file_name = np.load(f"SWAG_tune_lr/128_0.003_5e-3_0.001_85_25_20/val_results_30/val_eval_swag-46.npz")
residuals = (file_name["targets"]  - np.expand_dims(file_name["prediction_mean"], 1))
plt.hist(residuals, bins=np.arange(-0.25, 0.26, 0.01), label="20 epochs");

file_name = np.load(f"SWAG_tune_lr/128_0.003_5e-3_0.001_85_25_20/val_results_30/val_eval_swag-56.npz")
residuals = (file_name["targets"]  - np.expand_dims(file_name["prediction_mean"], 1))
plt.hist(residuals, bins=np.arange(-0.25, 0.26, 0.01), alpha=0.5, label="30 epochs");
plt.legend()
```

```python
file_name = f"SWAG_tune_lr/128_0.003_5e-3_0.001_85_25_20/val_results_100/val_eval_swag-56_residual_summary_100.csv"
df = pd.read_csv(file_name)
plt.plot(np.arange(len(df))+1, df.ens_of_std, label=f"30 epochs")
plt.legend()
plt.xlabel("Samples")
plt.ylabel("OF STD of N samples")
plt.xticks(np.arange(0, 110, 10))
plt.grid(True)
plt.ylim([0.073, 0.080])
```

```python
file_name = f"SWAG_tune_lr/128_0.003_5e-3_0.001_85_25_20/val_results_100/val_eval_swag-56_residual_summary_100.csv"
df = pd.read_csv(file_name)
plt.plot(np.arange(len(df))+1, df.ens_rms, label=f"30 epochs")
plt.legend()
plt.xlabel("Samples")
plt.ylabel("Ensemble RMS of N samples")
plt.xticks(np.arange(0, 110, 10))
plt.grid(True)
plt.ylim([0.098, 0.102])
```
