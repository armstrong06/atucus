```python
import numpy as np
import pandas as pd
import h5py
import matplotlib.pyplot as plt
```

```python
data_path = "s_resampled"
phase_type = "S"
```

```python
def load_data(data_path, split_type):
    df = pd.read_csv(f"{data_path}/uuss_{split_type}.csv")
    #df = df.drop("Unnamed: 0", axis=1)

    file = h5py.File(f"{data_path}/uuss_{split_type}.h5", "r")
    X = file["X"][:]
    file.close()

    df.head()
    
    return df, X
```

```python
def compute_snrs(X, phase_type="P", azims=None, noise_duration=1, signal_duration=0.5, buffer=15):
    """
    X - times series data
    phase_type - P or S
    azims - Back azimuth values for S
    noise_duration - Seconds to use for noise calculation
    signal_duration - Seconds after P arrival to use for signal calculation
    p_buffer: Number of samples prior to the P arrival to start the signal window
    """
    arrival_sample = X.shape[1]//2

    signal_ind_0 = int(arrival_sample - buffer)
    signal_ind_1 = int(arrival_sample + signal_duration*100)
    print(signal_ind_0, signal_ind_1)
    noise_ind_1 = int(signal_ind_0 - 1)
    noise_ind_0 = int(noise_ind_1 - noise_duration*100)
    print(noise_ind_0, noise_ind_1)

    snrs = np.full(X.shape[0], -999.9)
    
    if (phase_type != "P" and azims is None):
        print("Warning azims is none - don't use these SNRs")
    elif (azims is not None):
        print("Rotating to transverse")
            
    for i, x in enumerate(X):

        if (phase_type == "P"):
            x_signal = x[signal_ind_0:signal_ind_1] # Index 2 should be Z
            x_noise = x[noise_ind_0:noise_ind_1]
            x_signal = x_signal - np.mean(x_signal)
            x_noise = x_noise - np.mean(x_noise)

        else:
            if (azims is None):
                alpha = 0
            else:
                alpha = azims[i]*np.pi/180
            X_e = x[:, 0]
            X_n = x[:, 1]

            n_signal = X_n[signal_ind_0:signal_ind_1] # Index 2 should be Z
            n_noise = X_n[noise_ind_0:noise_ind_1]
            e_signal = X_e[signal_ind_0:signal_ind_1]
            e_noise = X_e[noise_ind_0:noise_ind_1]

            # transverse signal
            t_signal = -n_signal*np.sin(alpha) + e_signal*np.cos(alpha)
            t_noise  = -n_noise*np.sin(alpha)  + e_noise*np.cos(alpha)
            x_signal = t_signal - np.mean(t_signal)
            x_noise = t_noise - np.mean(t_noise)

        s2 = np.multiply(x_signal, x_signal) # element-wise
        n2 = np.multiply(x_noise, x_noise) # element-wise
        snr = np.mean(s2)/np.mean(n2) # Ratio of exepctations
        snr = 10*np.log10(snr)
        snrs[i] = snr
    return snrs
```

```python
def plot_snr_hist(meta_df, snrs, title=None):
    zero_weights = meta_df[meta_df["pick_quality"]==1].index
    one_weights = meta_df[meta_df["pick_quality"]==0.75].index
    two_weights = meta_df[meta_df["pick_quality"]==0.5].index
    assert len(zero_weights) + len(one_weights) + len(two_weights) == len(meta_df)
    plt.hist([snrs[zero_weights], snrs[one_weights], snrs[two_weights]], np.arange(-20, 80, 1), stacked=True,
             label=["zero", "one", "two"], edgecolor="k", linewidth=0.5);
    plt.title(title)
    plt.legend()
```

```python
test_df, test_X = load_data(data_path, "test_fewerhist")
test_azims = test_df.source_receiver_azimuth
test_fewerhist_snrs = compute_snrs(test_X, phase_type, test_azims, noise_duration=3, signal_duration=3, buffer=25)
plot_snr_hist(test_df, test_fewerhist_snrs, title="Test (fewerhist) SNRs")
```

```python
val_df, val_X = load_data(data_path, "validation")
val_azims = val_df.source_receiver_azimuth
validation_snrs = compute_snrs(val_X, phase_type, val_azims, noise_duration=3, signal_duration=3, buffer=25)
plot_snr_hist(val_df, validation_snrs, title="Validation SNRs")
```

```python
NGB_df, NGB_X = load_data(data_path, "NGB")
NGB_azims = NGB_df.source_receiver_azimuth
NGB_snrs = compute_snrs(NGB_X, phase_type, NGB_azims, noise_duration=3, signal_duration=3, buffer=25)
plot_snr_hist(NGB_df, NGB_snrs, title="NGB SNRs")
```
